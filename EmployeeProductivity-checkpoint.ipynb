<h1 class="custom-text"><b>Correlation Analysis</b></h1>
<h3 class="custom-text">Correlation is a statistical measure that quantifies the relationship between two variables.</h3>
<h3 class="custom-text">It is often represented by a correlation coefficient, such as the Pearson correlation coefficient, which can range from -1 to 1. Here is what the different values of the correlation coefficient mean:</h3>
<h3 class="custom-text"><b>Positive Correlation:</b> A correlation coefficient between 0 and 1 indicates a positive relationship between two variables. In this case, as one variable increases, the other tends to increase as well.</h3>
<h3 class="custom-text"><b>Negative Correlation:</b> A correlation coefficient between -1 and 0 indicates a negative relationship between two variables. As one variable increases, the other tends to decrease.</h3>
<h3 class="custom-text"><b>No Correlation:</b> A correlation coefficient close to 0 indicates little to no linear relationship between the variables. Changes in one variable do not predict changes in the other.</h3>
<h3 class="custom-text"><b>Perfect Correlation:</b> A correlation coefficient of 1 (positive) or -1 (negative) indicates a perfect linear relationship between the variables. In a perfect positive correlation, a change in one variable corresponds exactly with a change in the other in the same direction. In a perfect negative correlation, a change in one variable corresponds exactly with a change in the other in the opposite direction.</h3>



```python
correlation_matrix = df.corr() #corr() is use to calculate correlation coefficients 
correlation_matrix
```

<h1><b> Comparison of Productivity Level Across Departments and Position</b></h1> 

```python
# Create a figure and axes for the plot
fig, ax = plt.subplots(figsize=(10, 6))

# Comparison using a beeswarm plot
sns.swarmplot(x='Department', y='Productivity (%)', hue='Position', data=df)


# Add labels and title
plt.xlabel('Department')  # Label x-axis
plt.ylabel('Productivity (%)')  # Label y-axis
plt.legend(title='Position')  # Create legend
plt.title("Evaluating Productivity Across Departments and Positions")  # Title the plot

# Display plot
plt.show()

```

<h1><b>Projects Completed by Different Department</b></h1>

```python
# Group the data by department and position and sum the projects completed
projects_completed = df.groupby(['Department', 'Position'])['Projects Completed'].sum().reset_index()

# Create a barplot of projects completed by each position in each department
plt.figure(figsize=(12, 8))
sns.barplot(x='Department', y='Projects Completed', hue='Position', data=projects_completed)
plt.title('Projects Completed by Each Departmental Position')
plt.xlabel('Department')
plt.ylabel('Projects Completed')
plt.legend(title='Position')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,6))

sns.set_style('white')

#ax.scatter(df['Age'], df['Productivity (%)'])
sns.barplot(x='Satisfaction Rate (%)', y='Department', data=df, errcolor='none')
plt.xlabel('Satisfaction Rate (%)')
plt.ylabel('Department')
plt.title('Satisfaction Rate (%) by Different Department')

plt.show()
```

<h3><b>Feedback Score by Different Department</b></h3>

```python
fig, ax = plt.subplots(figsize=(10,6))

sns.set_style('white')

#ax.scatter(df['Age'], df['Productivity (%)'])
sns.barplot(x='Feedback Score', y='Department', data=df, errcolor='none')
plt.xlabel('Feedback Score')
plt.ylabel('Department')
plt.title('Feedback Score by Different Department')

plt.show()
```

<h3><b>Salary Rate</b></h3>

```python
# Group the data by department and calculate the minimum and maximum salaries
department_salaries = df.groupby('Department')['Salary'].agg(['min', 'max']).reset_index()

# Group the data by position and calculate the minimum and maximum salaries
position_salaries = df.groupby('Position')['Salary'].agg(['min', 'max']).reset_index()

# Plotting minimum and maximum salaries across departments
fig, ax = plt.subplots(2, 1, figsize=(12, 10))

# Minimum salaries
sns.barplot(x='Department', y='min', data=department_salaries, ax=ax[0], palette='Blues')
ax[0].set_title('Minimum Salary Across Departments')
ax[0].set_xlabel('Department')
ax[0].set_ylabel('Minimum Salary')
ax[0].tick_params(axis='x', rotation=45)

# Maximum salaries
sns.barplot(x='Department', y='max', data=department_salaries, ax=ax[1], palette='Reds')
ax[1].set_title('Maximum Salary Across Departments')
ax[1].set_xlabel('Department')
ax[1].set_ylabel('Maximum Salary')
ax[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()


```

<h3><b>Minimum and Maximum Salary Across Positions</b></h3>

```python
# Plotting minimum and maximum salaries across positions
fig, ax = plt.subplots(2, 1, figsize=(12, 10))

# Minimum salaries
sns.barplot(x='Position', y='min', data=position_salaries, ax=ax[0], palette='Blues')
ax[0].set_title('Minimum Salary Across Positions')
ax[0].set_xlabel('Position')
ax[0].set_ylabel('Minimum Salary')
ax[0].tick_params(axis='x', rotation=45)

# Maximum salaries
sns.barplot(x='Position', y='max', data=position_salaries, ax=ax[1], palette='Reds')
ax[1].set_title('Maximum Salary Across Positions')
ax[1].set_xlabel('Position')
ax[1].set_ylabel('Maximum Salary')
ax[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```
